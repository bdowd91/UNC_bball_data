```python
url = 'https://www.espn.com/mens-college-basketball/team/stats/_/id/153'
```

```python
tables = pd.read_html(url)
tables
```

```python
type(tables)
```

```python
stat_totals = tables[3]
stat_totals
```

```python
stat_avgs = tables[1]
```

```python
stat_totals.reset_index
```

```python
stat_totals_clean = pd.DataFrame(stat_totals)
stat_totals_clean
```

```python
PPG = stat_totals_clean["PTS"]/22
RBG = stat_totals_clean["REB"]/22
stat_totals_clean["PPG"] = PPG
stat_totals_clean["RBG"] = RBG

stat_totals_clean.head()
```
